```python
K=[1,2,5,10,50,100,200]
errTrain=[]
errVali=[]
for i,k in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr,k);
    Yhat = learner.predict(Xva)
    errTrain.append(learner.err(Xtr,Ytr))
    errVali.append(learner.err(Xva,Yva))
plt.semilogx(K,errTrain,'r',K,errVali,'g')
plt.show()
```
